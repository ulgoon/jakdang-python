```python
old_list = [1,2,3,4,5]
new_list = []
for item in old_list:
    if item % 2 == 0:
        new_list.append(item * 2)
        
new_list
```

```python
num_list = [cat for cat in range(1,100+1)]
new_list = []

for cat in num_list:
    if cat % 15 == 0:
        new_list.append("fizzbuzz")
    else:
        if cat % 3 == 0:
            new_list.append("fizz")
        else:
            if cat % 5 == 0:
                new_list.append("buzz")
            else:
                new_list.append(cat)
        
new_list
```

```python
new_list = ["fizzbuzz" if i%15==0 else "fizz" if i%3==0 else "buzz" if i%5==0 else i for i in num_list]
new_list
```
