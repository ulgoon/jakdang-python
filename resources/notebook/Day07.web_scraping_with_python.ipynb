```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.rottentomatoes.com/"
response = requests.get(url)
```

```python
response
```

```python
html = response.text
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
table = soup.find("table", attrs={"id":"Top-Box-Office"})
table
```

```python
all_tr = table.find_all("tr")
all_tr
```

```python
for tr in all_tr:
    all_td = tr.find_all("td")
    score = all_td[0].find("span", attrs={"class":"tMeterScore"}).text
    movie_name = all_td[1].a.text
    box_office = all_td[2].a.text.strip()
    print(score, movie_name, box_office)
```

```python
url = "https://www.naver.com/"
response = requests.get(url)
```

```python
html = response.text
html
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
ul = soup.find("ul", attrs={"class":"ah_l"})
```

```python
all_a = ul.find_all("a")
all_a
```

```python
for a in all_a:
    rank = a.find("span", attrs={"class":"ah_r"}).text
    keyword = a.find("span", attrs={"class":"ah_k"}).text
    print(rank, keyword)
    
```
